```python
df = pd.read_csv('/kaggle/input/cyberbullying-and-harassment-detection-using-ml/Approach to Social Media Cyberbullying and Harassment Detection Using Advanced Machine Learning.csv')
df.head()
```

```python
df.info()
```

```python
df['Label'].value_counts()
```

```python
X = df['Text']
Y = df['Label']
len(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.4, random_state= 22)
```

```python
len(X_train)
```

```python
len(Y_train)
```

```python
print("NaN in X_train:", X_train.isnull().any().sum())
print("NaN in Y_train:", Y_train.isnull().any().sum())
```

```python
# To drop rows with NaN values
X_train = X_train.dropna()
Y_train = Y_train.dropna()
```

```python
# Assuming Y_train has the missing sample
Y_train = Y_train.dropna()
X_train = X_train.loc[Y_train.index]
```

```python
print("X_train shape:", X_train.shape)
print("Y_train shape:", Y_train.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# !pip install pandas
import pandas as pd
# !pip install scikit-learn
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB, ComplementNB
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
pipeMNB = Pipeline([('tfidf', TfidfVectorizer()),('clf', MultinomialNB())])
pipeCNB = Pipeline([('tfidf', TfidfVectorizer()),('clf', ComplementNB())])
pipeSVC = Pipeline([('tfidf', TfidfVectorizer()),('clf', LinearSVC())])
```

```python
pipeMNB.fit(X_train, Y_train)
predictMNB = pipeMNB.predict(X_test)
print(f"MNB: {accuracy_score(Y_test, predictMNB):.2f}")
pipeCNB.fit(X_train, Y_train)
predictCNB = pipeCNB.predict(X_test)
print(f"CNB: {accuracy_score(Y_test, predictCNB):.2f}")
pipeSVC.fit(X_train, Y_train)
predictSVC = pipeSVC.predict(X_test)
print(f"SVC: {accuracy_score(Y_test, predictSVC):.2f}")
# Convert both Y_test and predictMNB to strings
Y_test = Y_test.astype(str)
predictMNB = predictMNB.astype(str)
```

```python
pipeMNB = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words='english', ngram_range=(1,3))),
    ('clf', MultinomialNB())
])
```

```python
pipeMNB.fit(X_train, Y_train)
predictMNB = pipeMNB.predict(X_test)
print(f"MNB: {accuracy_score(Y_test, predictMNB):.2f}")
pipeCNB.fit(X_train, Y_train)
predictCNB = pipeCNB.predict(X_test)
print(f"CNB: {accuracy_score(Y_test, predictCNB):.2f}")
pipeSVC.fit(X_train, Y_train)
predictSVC = pipeSVC.predict(X_test)
print(f"SVC: {accuracy_score(Y_test, predictSVC):.2f}")
```
